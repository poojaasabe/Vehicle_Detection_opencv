```python

import numpy as np
import cv2

import os
import re
from os.path import isfile, join
%matplotlib inline
from matplotlib import pyplot as plt

np.random.seed(42)
```

```python
# get file names of the frames
col_frames = os.listdir('frames/')

print(col_frames)

```

```python
# sort file names
col_frames.sort(key=lambda f: int(re.sub('\D','', f)))

print(col_frames)
```

```python
# empty list to store the frames
col_images=[]

for i in col_frames:
    # read the frames
    img = cv2.imread('frames/'+i)
    # append the frames to the list
    col_images.append(img)
```

```python
# plot 13th frame
i = 13

for frame in [i, i+1]:
    plt.imshow(cv2.cvtColor(col_images[frame], cv2.COLOR_BGR2RGB))
    plt.title("frame: "+str(frame))
    plt.show()
```

```python
# convert the frames to grayscale
grayA = cv2.cvtColor(col_images[i], cv2.COLOR_BGR2GRAY)
grayB = cv2.cvtColor(col_images[i+1], cv2.COLOR_BGR2GRAY)

# plot the image after frame differencing
plt.imshow(cv2.absdiff(grayB, grayA), cmap = 'gray')
plt.show()
```

```python

blurred = cv2.GaussianBlur(cv2.absdiff(grayB, grayA), (11,11), 0)
plt.imshow(blurred)
plt.show()
```

```python
diff_image = cv2.absdiff(grayB, grayA)

# perform image thresholding
ret, thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# plot image after thresholding
plt.imshow(thresh, cmap = 'gray')
plt.show()
```

```python
# find contours

(cnts, _) = cv2.findContours(thresh.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
valid_cntrs = []
sample=col_images[13].copy()
for cntr in cnts:
    x,y,w,h = cv2.boundingRect(cntr)
    if (x <= 200) & (y >= 80) & (cv2.contourArea(cntr) >= 25) :
        valid_cntrs.append(cntr)
        cv2.rectangle(sample,(x,y),(x+w,y+h),(0,255,0),2)

# count of discovered contours        
len(valid_cntrs)
plt.imshow(sample)
```

```python
# font style
font = cv2.FONT_HERSHEY_SIMPLEX

# empty list to store the frames
col_images=[]

for i in col_frames:
    # read the frames
    img = cv2.imread('frames/'+i)
    # append the frames to the list
    col_images.append(img)


for i in range(len(col_images)-1):
    sample = col_images[i].copy()
    
    # frame differencing
    grayA = cv2.cvtColor(col_images[i], cv2.COLOR_BGR2GRAY)
    grayB = cv2.cvtColor(col_images[i+1], cv2.COLOR_BGR2GRAY)
    diff_image = cv2.absdiff(grayB, grayA)
    
    #blurring
    blurred = cv2.GaussianBlur(cv2.absdiff(grayB, grayA), (11,11), 0)
    
    # image thresholding
    ret, thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    
    
    # find contours
    (cnts, _) = cv2.findContours(thresh.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    
    # shortlist contours appearing in the detection zone
    valid_cntrs = []
    
    #loop over contours and form rectangle shaped contours
    for cntr in cnts:
        x,y,w,h = cv2.boundingRect(cntr)
        if (x <= 200) & (y >= 60) & (cv2.contourArea(cntr) >= 25) :
            
            valid_cntrs.append(cntr)
            cv2.rectangle(sample,(x,y),(x+w,y+h),(0,255,0),2)
    
    #cv2.imwrite('cnt/'+str(i)+'.png',sample)        
    # add contours to original frame
    
    cv2.putText(sample, "vehicles detected: " + str(len(valid_cntrs)), (55, 15), font, 0.6, (0, 180, 0), 2 )
    #print(i)
    #plt.imshow(sample)
    #plt.show()
    cv2.imwrite('cnt/'+str(i)+'.png',sample)  
```

```python
# specify video name
pathOut = 'vehicle_detection_v3.mp4'

# specify frames per second
fps = 14.0
#empty array
frame_array = []
files = [f for f in os.listdir('cnt/') if isfile(join('cnt/', f))]
```

```python
#sort the images
files.sort(key=lambda f: int(re.sub('\D', '', f)))
print(files)
```

```python


for i in range(len(files)):
    filename= 'cnt/'+files[i]
    
    #read frames
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    
    #inserting the frames into an image array
    frame_array.append(img)
```

```python
#write the video
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)

for i in range(len(frame_array)):
    # writing to a image array
    out.write(frame_array[i])

out.release()
```
